```python
import pycountry

def country_to_alpha_3(country_name):
    try:
      if country_name == 'Bolivia': 
        country_name = 'Bolivia, Plurinational State of'
      if country_name == 'Egypt, Arab Rep.': 
        country_name = 'Egypt'
      if country_name == 'United States of America': 
        country_name = 'United States'
      if country_name == 'The Commonwealth of Australia': 
        country_name = 'Australia'
      if country_name == 'The Russian Federation': 
        country_name = 'Russian Federation'
      return pycountry.countries.get(name=country_name).alpha_3
    except AttributeError:
      return "Unknown country"

print(country_to_alpha_3('The Russian Federation'))
```

```python
iso_code = []

for i in range(1,len(data)): 
  iso_code.append(country_to_alpha_3(data[i][0]))

df['iso_codes'] = iso_code
print(df)
```

```python
import plotly.express as px
fig = px.choropleth(df, locations="iso_codes",
                    color="Income group", # lifeExp is a column of gapminder
                    hover_name="\ufeffCountry", # column to add to hover information
                    )
fig.show()
```

```python
import plotly.io as pio
fh = open('income.html', 'w') # this just creates a blank file, so that the write_html() function can write to this file later
fh.close()

pio.write_html(fig, file='income.html', auto_open=True, full_html=False) # preferred method
```

```python
#choropleth of agricultural sector over time - EXPORT
fh = open('Wto_agricultural.csv', encoding='cp1252')
reader = csv.reader(fh)
data = list(reader)
fh.close()
for i in range(1,len(data)): 
  data[i][2] = int(data[i][2])
  data[i][3] = float(data[i][3])
agri_df = pd.DataFrame(data[1:], columns=data[0])
#print(agri_df)

iso_code = []

for i in range(1,len(data)): 
  iso_code.append(country_to_alpha_3(data[i][0]))

agri_df['iso_codes'] = iso_code
print(agri_df)
```

```python
fig = px.choropleth(agri_df, locations='iso_codes', range_color=[0, 100000], color='Value', hover_name='Reporting Economy',
                    animation_frame= 'Year', title="Choropleth of Country Export Values of Agricultural Product in million(s) USD")
fig.show()
```

```python
import plotly.io as pio
fh = open('agri_ex.html', 'w') 
fh.close()

pio.write_html(fig, file='agri_ex.html', auto_open=True, full_html=False)
```
