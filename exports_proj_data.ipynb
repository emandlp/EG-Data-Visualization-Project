```python
import pycountry

def country_to_alpha_3(country_name):
    try:
      if country_name == 'Bolivia': 
        country_name = 'Bolivia, Plurinational State of'
      if country_name == 'Egypt, Arab Rep.': 
        country_name = 'Egypt'
      if country_name == 'United States of America': 
        country_name = 'United States'
      if country_name == 'The Commonwealth of Australia': 
        country_name = 'Australia'
      if country_name == 'The Russian Federation': 
        country_name = 'Russian Federation'
      return pycountry.countries.get(name=country_name).alpha_3
    except AttributeError:
      return "Unknown country"

print(country_to_alpha_3('The Russian Federation'))
```
